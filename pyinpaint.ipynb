## Image Completion via Patch Statistics

Image completion is the task of filling an unknown region of an image such that the "completed" image is visually appealing. This notebook implement the image completion algorithm by _Kaiming He and Jian Sun_, [Statistics of Patch Offsets for Image Completion](http://research.microsoft.com/en-us/um/people/kahe/eccv12/index.html).  The paper itself is an excellent reference and is very readable.

The method in the paper is based on the observation that most natural images have a few characteristic offsets between similar regions of an image.  Given these _dominant offsets_ calculated from the known region, image completion is just selecting the "best" such offset for each pixel in the unknown region.  The unknown pixel then takes on the color of the pixel at that offset.  Since the number of dominant offsets is small, the search space of the labeling task is relative small and an approximate solution can be found quickly.

### Outline

1. Visualize known and unknown regions
1. Determine patch offsets
1. Create histogram and find top $K$ offsets
1. Find optimal inpainting solution via graph cuts

In this notebook, I do not focus on computational efficiency, particularly in step 2 when calculating the image offsets.

### A word of warning

Beware that a[m][n] means to index the m-th row and n-th column of the image, but the equivalent $x$,$y$ coordinate is $(n,m)$.  That is, I often think $a[x][y]$ would be the pixel at $(x,y)$, but that is really $a[y][x]$.

### Determine patch offsets

A patch is an $8\times 8$ sample of the image.  The patch $P(\mathbf{x})$ is the patch with upper-left hand pixel $\mathbf{x}$.  A patch with any overlap in the unknown region is considered unknown and will not be used to characterize patch statistics.

We want to characterize the typical offsets of similar patches in the known region.  Of course, a patch is most similar to itself, so we set a threshold $\tau$ as a minimum distance between compared patches.  The offset $\mathbf{s} = (u,v)$ is given by:

\begin{equation}
\mathbf{s(x)} = \arg\min_s \  ||P(\mathbf(x+s) - P(\mathbf(x)||^2  \ s.t \ |\mathbf{s}| > \tau
\end{equation}

All the distances are just sum over squares of RGB pixel values in the patches.

The histogram of offsets is then simply

\begin{equation}
h(u,v) = \sum_{x} \delta ( \mathbf{s(x)} = (u,v) )
\end{equation}


```python
rows = known.shape[0] # number of rows
cols = known.shape[1] # number of columns
tau = rows/15 # minimum distance between patches
D = 8 # patch size

def is_valid_patch(r,c,msk):
    ''' Returns true if patch is entirely in known region (and in image) '''
    return (r+D < rows) and (c + D < cols) and np.all(msk[r:r+8,c:c+8])

# quick lookup of whether patch is valid
valid_patch = np.zeros((rows,cols),dtype=np.int)
for r in range(0,rows):
    for c in range(0,cols):
        valid_patch[r,c] = is_valid_patch(r,c,knownMask)

@jit
def patch_overlap_fast(im, r1,c1,r2,c2):
    ''' Sum of squares distance over patch pixels '''
    channels = im.shape[2]
    res = 0.
    for dr in range(0,8):
        for dc in range(0,8):
           for channel in range(0,channels):
                tmp = im[r1+dr,c1+dc,channel] - im[r2+dr,c2+dc,channel]
                res += tmp * tmp
    return res

@jit
def min_offset_fast(r,c,tau,im,msk):
    ''' Find the offset of most similar patch at least tau away on pixel space '''
    rows = im.shape[0]
    cols = im.shape[1]
    res = (0,0)
    bestOverlap = 1e30
    bestCount = 0
    tau_sq = tau * tau
    for r1 in range(0,rows):
        for c1 in range(0,cols):
            if(msk[r1,c1]):
                dist = (r-r1) *(r-r1) + (c-c1) * (c-c1)
                if(dist >= tau_sq):
                    p = patch_overlap_fast(im,r,c,r1,c1)
                    if(p < bestOverlap):
                        bestOverlap = p
                        res = (r1-r,c1-c)
                        bestCount = 1
                    elif(p == bestOverlap):
                        # randomly replace equivalent offset
                        bestCount += 1
                        if(np.random.uniform() < 1./bestCount):
                            res = (r1 - r, c1 - c)
    return res

@jit
def patch_statistics(im,msk):
    '''
    Return offset histogram of patch overlaps
    im : RGB image
    msk : boolean mapping where msk[r][c] means that patch is valid/known
    '''
    rows = im.shape[0]
    cols = im.shape[1]
    res = np.zeros((rows*2-1,cols*2-1))
    for r in range(0,rows):
        for c in range(0,cols):
            if(msk[r,c]):
                s = min_offset_fast(r,c,tau,im,msk)
                res[s[0]+rows-1,s[1]+cols-1] += 1
    return res
```

### Find peak offsets

Given $h(u,v)$ above, the paper first smooths the histogriam with a gaussian filter to make it easier to find local maximum.  As discussed at the end, I found that random offsets were still ruining the image completion results, so I further remove small count bins after smoothing.  A dominant offset is one that is locally maximal in an $8 \times 8$ window centered around that bin in the histogram.

```python
smoothed_offsets = gaussian_filter(raw_offsets,sqrt(2))
# heuristic filter of "noise" offsets .. hand-wavey binomial argument here
smoothed_offsets[smoothed_offsets < 5] = 0
```

```python
@jit
def is_peak(offsets, r, c):
    if(offsets[r,c] == 0):
        return False
    for dr in range(-4,5):
        if(r + dr >= 0 and r + dr < offsets.shape[0]):
            for dc in range(-4,5):
                if(c+dc >= 0 and c + dc < offsets.shape[1]):
                    if(offsets[r,c] < offsets[r+dr,c+dc]):
                        return False
    return True

@jit
def dominant_offsets_helper(offsets):
    ''' Return K most dominant offsets from offsets matrix '''
    res = np.zeros((10000,3))
    idx = 0
    rows = offsets.shape[0]
    cols = offsets.shape[1]
    for r in range(0,rows):
        for c in range(0,cols):
            if(is_peak(offsets,r,c)):
                res[idx,0] = r
                res[idx,1] = c
                res[idx,2] = offsets[r,c]
                idx += 1
    return res[:idx,:]
                
def dominant_offsets(im, offsets, K = 60):
    ''' Collect dominate offsets from image, adding in default 0,0 offset '''
    tmp = dominant_offsets_helper(offsets)
    dominant = np.array(sorted(tmp, key = lambda x: x[-1])[::-1])[:K,:]
    dominant_offsets = dominant[:,:2]
    dominant_counts = dominant[:,-1]
    res_offsets = np.concatenate((np.array([[0,0]]),dominant_offsets - [im.shape[0] - 1,im.shape[1] - 1]))
    res_counts = np.concatenate((np.array([0]), dominant_counts))
    return res_offsets, res_counts
```

### Prepare Energy Minimization

We now want to find the best offset for unknown pixels.  That is, we seek $L(x)$, the offset label or index for pixel $x$.  "Best" is posed via an energy minization problem:

\begin{equation}
E(L) = \sum_{x\in\Omega} E_d(L(x)) + \sum_{x,x' | x \in \Omega, x' \in \Omega} E_s(L(x),L(x'))
\end{equation}

Where $\Omega$ is the unknown region and $(x,x')$ are 4-connected neighbors.   The single site energy $E_d(L(x))$ is 0 if that label is valid and infinite if the label is invalid for the pixel.  If offset $L(x) = i$, that means we use the $i$-th offset to find the value for pixel $x$.  That is, $x = x + o_i$.  If $x + o_i$ is not in the known region, it is an invalid setting.  For pixels on the boundary (known region), the only valid offset is the $(0,0)$ offset.

The two site energy is
\begin{equation}
E_s(L(x),L(x')) = \lVert I(x + o_{L(x)}) - I(x + o_{L(x')})\rVert^2 + \lVert I(x' + o_{L(x)})- I(x' + o_{L(x')})\rVert^2
\end{equation}

That is, since $x,x'$ are neighbors in the source image, different labels/offsets in the final image can potentially create a seam. $E_s$ penalizes these seams in favor of smoothness.

Consult the pygco source for details on how these energies are given to the graph-cut algorithm that minimizes $E(L)$.

```python
@jit
def prep_single_site_energies(knownMask, offsets):
    ''' Prepares $E_d$ as described above '''
    # energy is 0 if offset is valid
    # energy is infinity if invalid
    rows = knownMask.shape[0]
    cols = knownMask.shape[1]
    num_labels = len(offsets)
    res = np.zeros((rows,cols,num_labels), dtype = np.int32)
    myinf = 10000000
    for r in range(rows):
        for c in range(cols):
            for k in range(num_labels):
                # if pixel is known, only allow 0 offset
                if(knownMask[r,c]):
                    if(k == 0):
                        res[r,c,k] = 0
                    else:
                        res[r,c,k] = myinf
                else: # unknown pixel, only allow valid offsets
                    r1 = r + offsets[k,0]
                    c1 = c + offsets[k,1]
                    if(r1 >= 0 and r1 < rows and c1 >= 0 and c1 < cols and knownMask[r1,c1]):
                        res[r,c,k] = 0
                    else:
                        res[r,c,k] = myinf
    return res
```

```python
unary_costs = prep_single_site_energies(knownMask,offsets)
```

```python
labels = cut_inpaint(unary_costs.astype(np.int32), offsets.astype(np.int32), ( known * 255).astype(np.int32), knownMask.astype(np.int32), n_iter = -1, algorithm = 'swap', verbosity = 1)
```

```python
final = known.copy()
for r in xrange(labels.shape[0]):
    for c in xrange(labels.shape[1]):
        o = offsets[labels[r,c],:]
        r1 = r + o[0]
        c1 = c + o[1]
        if(r1 >= 0 and r1 < known.shape[0] and c1 >= 0 and c1 < known.shape[1]):
            final[r,c,:] = known[r1, c1,:]
        else:
            print r,c, r1,c1
```

### Final Comments

For the most part, the implementation followed straightforwardly from the paper.  However, the graph cut algorithm kept getting stuck in local minima with poor visual results.  I blame the choice of $K=60$ dominant offsets is too generic.  If there are really say 10 dominant offsets, the extra 50 are noise that muck up the energy landscape with local minima.  There may be something I'm still missing from the paper, but I do not see how the authors address this issue.  

My solution for now is to filter out low-count bins after smoothing.  If offsets were all equally likely, then the histogram is multinomial with an extremely low probability to have even more than a handful of counts.

Long term, I suspect the better solution is to look at the CDF of the histogram and find where the inflection occurs.  As seen below the counts of the dominant offset peaks falls off fairly quickly.

```python
offsets, counts = dominant_offsets(known,gaussian_filter(raw_offsets,sqrt(2)), K = 60)
plot(np.cumsum(counts[1:]))
ylabel('Cumulative Count')
xlabel('Dominant Offset Index');
counts[1:20]
```
