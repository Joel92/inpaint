## Image Completion via Patch Statistics

The main goal is to implement the image completion algorithm by _Kaiming He and Jian Sun_, [Statistics of Patch Offsets for Image Completion](http://research.microsoft.com/en-us/um/people/kahe/eccv12/index.html).  The steps below are the bare minimum needed.  Once this is working on the test image, I will go back and use smarter techniques for the patch representation and offset calculations.

1. Visualize known and unknown regions
1. Determine patch offsets
1. Create histogram and find top $K=60$ offsets
1. Find optimal inpainting solution using graph cut library

#### A word of warning

Beware that a[m][n] means to index the m-th row and n-th column of the image.  But the equivalent $x$,$y$ coordinate is $(n,m)$.  That is, I often think $a[x][y]$ would be the pixel at $(x,y)$, but that is really $a[y][x]$.

#### Determine patch offsets

A patch is an $8\times 8$ sample of the image.  The patch $P(\mathbf{x})$ is the patch with upper-left hand pixel $\mathbf{x}$.  A patch with any overlap in the unknown region is considered unkown and will not be used to characterize patch statistics.

We want to characterize the typical offsets of similar patches in the known region.  Of course, a patch is most similar to itself, so we set a threshold $\tau$ as a minimum distance between compared patches.  The offset $\mathbf{s}(u,v)$ is given by:

\begin{equation}
\mathbf{s(x)} = \arg\min_s \  ||P(\mathbf(x+s) - P(\mathbf(x)||^2  \ s.t \ |\mathbf{s}| > \tau
\end{equation}

For now, all the distances will just be sum over squares of all matrix elements.  

```python
tau = 200/15 # minimum distance between patches
M = known.shape[0] # number of rows
N = known.shape[1] # number of columns

valid_patch = np.zeros((M,N),dtype=np.int)

def is_valid_patch(m,n,msk):
    ''' Returns true if patch is entirely in known region (and in image) '''
    return (m+D < M) and (n + D < N) and np.all(msk[m:m+8,n:n+8])

for m in range(0,M):
    for n in range(0,N):
        valid_patch[m,n] = is_valid_patch(m,n,knownMask)

def patch_overlap(p1,p2):
    ''' Returns the distances (squared) between two patches '''
    return np.sum((p1 - p2) * (p1 - p2))

def distance_sq(x1,x2):
    ''' Return distance squared between 2-d tuples x1,x2 '''
    x = np.array(x1)-np.array(x2)
    return np.sum(x*x)

def iter_region(msk):
    ''' Iterate patch coordinates where msk is true '''
    M = msk.shape[0]
    N = msk.shape[1]
    for m in range(0,M):
        for n in range(0,N):
            if(is_valid_patch(m,n,msk)):
                yield (m,n)

def min_offset(m,n,tau,im):
    ''' Returns offset to most similar patch at least tau away '''
    res = (0,0)
    bestOverlap = 1e30
    tau_sq = tau * tau
    for m1,n1 in iter_region(knownMask):
        dist = distance_sq((m,n),(m1,n1))
        if(dist >= tau_sq):
            p = patch_overlap(im[m:(m+8),n:(n+8)],im[m1:(m1+8),n1:(n1+8)] )
            if(p < bestOverlap):
                bestOverlap = p
                res = (m1-m,n1-n)
    return res

#@jit(double(double[:,:,:], int_,int_,int_,int_))
@autojit
def patch_overlap_fast(im, m1,n1,m2,n2):
    C = im.shape[2]
    res = 0.
    for dm in range(0,8):
        for dn in range(0,8):
           for c in range(0,C):
                tmp = im[m1+dm,n1+dn,C] - im[m2+dm,n2+dn,C]
                res += tmp * tmp
    return res

#@jit(double[:,:](int_,int_,int_,double[:,:,:],int_[:,:]))
@autojit
def min_offset_fast(m,n,tau,im,msk):
    M = im.shape[0]
    N = im.shape[1]
    res = (0,0)
    bestOverlap = 1e30
    tau_sq = tau * tau
    for m1 in range(0,M):
        for n1 in range(0,N):
            if(msk[m1,n1]):
                dist = (m-m1) *(m-m1) + (n-n1) * (n-n1)
                if(dist >= tau_sq):
                    p = patch_overlap_fast(im,m,n,m1,n1)
                    if(p < bestOverlap):
                        bestOverlap = p
                        res = (m1-m,n1-n)
    return res

@autojit
def patch_statistics(im,msk):
    M = im.shape[0]
    N = im.shape[1]
    res = np.zeros((M*2,N*2))
    for m in range(0,M):
        for n in range(0,N):
            if(msk[m,n]):
                s = min_offset_fast(m,n,tau,im,msk)
                res[s[0]+M,s[1]+N] += 1
    return res
                
```

### Find peak offsets

```python
offsets[offsets < 2] = 0
smoothed_offsets = gaussian_filter(offsets,sqrt(2))
```

```python
@autojit
def is_peak(offsets, m, n):
    if(offsets[m,n] == 0):
        return False
    for dm in range(-4,5):
        if(m + dm >= 0 and m + dm < offsets.shape[0]):
            for dn in range(-4,5):
                if(n+dn >= 0 and n + dn < offsets.shape[1]):
                    if(offsets[m,n] < offsets[m+dm,n+dn]):
                        return False
    return True

@autojit
def dominant_offsets_helper(offsets):
    ''' Return K most dominant offsets from offsets matrix '''
    res = np.zeros((10000,3))
    idx = 0
    M = offsets.shape[0]
    N = offsets.shape[1]
    for m in range(0,M):
        for n in range(0,N):
            if(is_peak(offsets,m,n)):
                res[idx,0] = m
                res[idx,1] = n
                res[idx,2] = offsets[m,n]
                idx += 1
    return res[:idx,:]
                
def dominant_offsets(im, offsets, K = 60):
    ''' Collect dominate offsets from image, adding in default 0,0 offset '''
    res = dominant_offsets_helper(offsets)
    res = np.array(sorted(res, key = lambda x: x[-1])[::-1])[:K,0:2]
    return np.concatenate((np.array([[0,0]]),res - [im.shape[0],im.shape[1]]))
```

### Prepare Energy Minimization

We now want to find the best offset for unknown pixels.  That is, we seek $L(x)$, the offset label or index for pixel $x$.  "Best" is posed via an energy minization problem:

\begin{equation}
E(L) = \sum_{x\in\Omega} E_d(L(x)) + \sum_{x,x' | x \in \Omega, x' \in \Omega} E_s(L(x),L(x'))
\end{equation}

Where $\Omega$ is the unknown region and $(x,x')$ are 4-connected neighbors.   The single site energy $E_d(L(x))$ is 0 if that label is valid and infinite if the label is invalid for the pixel.  If offset $L(x) = i$, that means we use the $i$-th offset to find the value for pixel $x$.  That is, $x = x + o_i$.  If $x + o_i$ is not in the known region, it is an invalid setting.  For pixels on the boundary (known region), the only valid offset is the $(0,0)$ offset.

The two site energy is
\begin{equation}
E_s(L(x),L(x')) = \lVert I(x + o_{L(x)} - I(x + o_{L(x')}\rVert^2 + \lVert I(x' + o_{L(x)} - I(x' + o_{L(x')}\rVert^2
\end{equation}

That is, since $x,x'$ are neighbors in the source image, different labels/offsets in the final image can potentially create a seam. $E_s$ penalizes these seams in favor of smoothness.

Consult the pygco source for details on how these energies are given to the graph-cut algorithm that minimizes $E(L)$.

```python
@autojit
def prep_single_site_energies(knownMask, offsets):
    ''' Prepares $E_d$ as described above '''
    # energy is 0 if offset is valid
    # energy is infinity if invalid
    M = knownMask.shape[0]
    N = knownMask.shape[1]
    num_labels = len(offsets)
    res = np.zeros((M,N,num_labels), dtype = np.int32)
    myinf = np.iinfo(np.int32).max / 100
    for m in range(M):
        for n in range(N):
            for k in range(num_labels):
                # if pixel is known, only allow 0 offset
                if(knownMask[m,n]):
                    if(k == 0):
                        res[m,n,k] = 0
                    else:
                        res[m,n,k] = myinf
                else: # unknown pixel, only allow valid offsets
                    m1 = m + offsets[k,0]
                    n1 = n + offsets[k,1]
                    if(m1 >= 0 and m1 < M and n1 >= 0 and n1 < N and knownMask[m1,n1]):
                        res[m,n,k] = 0
                    else:
                        res[m,n,k] = myinf
    return res
```

```python
unary_costs = prep_single_site_energies(knownMask,offsets)
```

```python
labels = cut_inpaint(unary_costs.astype(np.int32), offsets.astype(np.int32), (known * 255).astype(np.int32), knownMask.astype(np.int32))
```
