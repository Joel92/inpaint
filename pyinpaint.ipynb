### YCbCr Colorspace Conversion

```python
rgb2ycbcr_mat = np.array([
    [0.299, 0.587, 0.114],
    [-0.168736,-0.331264,0.5],
    [0.5,-0.418688,-0.081312]
    ])

ycbcr_offsets = np.array([[16.],[128.],[128.]])
ycbcr_scales = np.array([[219.],[224.],[224.]])

def rgb2ycbcr(img):
    res = img.astype(np.float)/255.
    res = res.swapaxes(0,2)
    old_shape = res.shape
    res = res.reshape((3,-1))
    res = np.dot(rgb2ycbcr_mat,res)
    res = res * ycbcr_scales + ycbcr_offsets
    res.shape = old_shape
    res = np.swapaxes(res,2,0)
    return np.round(res).astype(np.uint8)

def ycbcr2rgb(img):
    res = img.astype(np.float)
    res = res.swapaxes(0,2)
    old_shape = res.shape
    res = res.reshape((3,-1))
    res = (res - ycbcr_offsets)/ycbcr_scales
    res = np.dot(np.linalg.inv(rgb2ycbcr_mat),res)
    res.shape = old_shape
    res = np.swapaxes(res,2,0)
    return np.round(res * 255.).astype(np.uint8)
```

#### Determine patch offsets

A patch is an $8\times 8$ sample of the image.  The patch $P(\mathbf{x})$ is the patch with upper-left hand pixel $\mathbf{x}$.  A patch with any overlap in the unknown region is considered unkown and will not be used to characterize patch statistics.

We want to characterize the typical offsets of similar patches in the known region.  Of course, a patch is most similar to itself, so we set a threshold $\tau$ as a minimum distance between compared patches.  The offset $\mathbf{s}(u,v)$ is given by:

\begin{equation}
\mathbf{s(x)} = \arg\min_s \  ||P(\mathbf(x+s) - P(\mathbf(x)||^2  \ s.t \ |\mathbf{s}| > \tau
\end{equation}

For now, all the distances will just be sum over squares of all matrix elements.  

```python
from numba import autojit, jit, double,int_

tau = 200/15 # minimum distance between patches
D = 8 # patch size
M = known.shape[0] # number of rows
N = known.shape[1] # number of columns

valid_patch = np.zeros((M,N),dtype=np.int)

def is_valid_patch(m,n,msk):
    ''' Returns true if patch is entirely in known region (and in image) '''
    return (m+D < M) and (n + D < N) and np.all(msk[m:m+8,n:n+8])

for m in range(0,M):
    for n in range(0,N):
        valid_patch[m,n] = is_valid_patch(m,n,knownMask)

def patch_overlap(p1,p2):
    ''' Returns the distances (squared) between two patches '''
    return np.sum((p1 - p2) * (p1 - p2))

def distance_sq(x1,x2):
    ''' Return distance squared between 2-d tuples x1,x2 '''
    x = np.array(x1)-np.array(x2)
    return np.sum(x*x)

def iter_region(msk):
    ''' Iterate patch coordinates where msk is true '''
    M = msk.shape[0]
    N = msk.shape[1]
    for m in range(0,M):
        for n in range(0,N):
            if(is_valid_patch(m,n,msk)):
                yield (m,n)

def min_offset(m,n,tau,im):
    ''' Returns offset to most similar patch at least tau away '''
    res = (0,0)
    bestOverlap = 1e30
    tau_sq = tau * tau
    for m1,n1 in iter_region(knownMask):
        dist = distance_sq((m,n),(m1,n1))
        if(dist >= tau_sq):
            p = patch_overlap(im[m:(m+8),n:(n+8)],im[m1:(m1+8),n1:(n1+8)] )
            if(p < bestOverlap):
                bestOverlap = p
                res = (m1-m,n1-n)
    return res

#@jit(double(double[:,:,:], int_,int_,int_,int_))
@autojit
def patch_overlap_fast(im, m1,n1,m2,n2):
    C = im.shape[2]
    res = 0.
    for dm in range(0,8):
        for dn in range(0,8):
           for c in range(0,C):
                tmp = im[m1+dm,n1+dn,C] - im[m2+dm,n2+dn,C]
                res += tmp * tmp
    return res

#@jit(double[:,:](int_,int_,int_,double[:,:,:],int_[:,:]))
@autojit
def min_offset_fast(m,n,tau,im,msk):
    M = im.shape[0]
    N = im.shape[1]
    res = (0,0)
    bestOverlap = 1e30
    tau_sq = tau * tau
    for m1 in range(0,M):
        for n1 in range(0,N):
            if(msk[m1,n1]):
                dist = (m-m1) *(m-m1) + (n-n1) * (n-n1)
                if(dist >= tau_sq):
                    p = patch_overlap_fast(im,m,n,m1,n1)
                    if(p < bestOverlap):
                        bestOverlap = p
                        res = (m1-m,n-n1)
    return res

@autojit
def patch_statistics(im,msk):
    M = im.shape[0]
    N = im.shape[1]
    res = np.zeros((M*2,N*2))
    for m in range(0,M):
        for n in range(0,N):
            if(msk[m,n]):
                s = min_offset_fast(m,n,tau,im,msk)
                res[s[0]+M,s[1]+N] += 1
    return res
                
```
