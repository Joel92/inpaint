#### A word of warning

Beware that a[m][n] means to index the m-th row and n-th column of the image.  But the equivalent $x$,$y$ coordinate is $(n,m)$.  That is, I often think $a[x][y]$ would be the pixel at $(x,y)$, but that is really $a[y][x]$.

```julia
groundTruth = fp_image(PyPlot.imread("img/dots_input.jpeg"))
unknown = fp_image(PyPlot.imread("img/dots_alpha.jpeg"))
unknown[unknown .< 100./255.] = 0 # fix dithering
unknownMask = unknown .== 0
knownMask = unknown .!= 0
known = copy(groundTruth)
slice(known,:,:,1)[unknownMask] = 1.
slice(known,:,:,2)[unknownMask] = 0.
slice(known,:,:,3)[unknownMask] = 0.

PyPlot.subplot(1,3,1)
PyPlot.imshow(groundTruth)
PyPlot.title("Ground Truth")
PyPlot.subplot(1,3,2)
PyPlot.imshow(unknown, cmap = PyPlot.get_cmap("gray"))
PyPlot.title("Unknown Region")
PyPlot.subplot(1,3,3)
PyPlot.imshow(known)
PyPlot.title("Overlay")
```

#### Determine patch offsets

A patch is an $8\times 8$ sample of the image.  The patch $P(\mathbf{x})$ is the patch with upper-left hand pixel $\mathbf{x}$.  A patch with any overlap in the unknown region is considered unkown and will not be used to characterize patch statistics.

We want to characterize the typical offsets of similar patches in the known region.  Of course, a patch is most similar to itself, so we set a threshold $\tau$ as a minimum distance between compared patches.  The offset $\mathbf{s}(u,v)$ is given by:

\begin{equation}
\mathbf{s(x)} = \arg\min_s \  ||P(\mathbf(x+s) - P(\mathbf(x)||^2  \ s.t \ |\mathbf{s}| > \tau
\end{equation}
