```python
@jit
def is_peak(offsets, r, c):
    if(offsets[r,c] == 0):
        return False
    for dr in range(-4,5):
        if(r + dr >= 0 and r + dr < offsets.shape[0]):
            for dc in range(-4,5):
                if(c+dc >= 0 and c + dc < offsets.shape[1]):
                    if(offsets[r,c] < offsets[r+dr,c+dc]):
                        return False
    return True

@jit
def dominant_offsets_helper(offsets):
    ''' Return K most dominant offsets from offsets matrix '''
    res = np.zeros((10000,3))
    idx = 0
    rows = offsets.shape[0]
    cols = offsets.shape[1]
    for r in range(0,rows):
        for c in range(0,cols):
            if(is_peak(offsets,r,c)):
                res[idx,0] = r
                res[idx,1] = c
                res[idx,2] = offsets[r,c]
                idx += 1
    return res[:idx,:]
                
def dominant_offsets(im, offsets, K = 60):
    ''' Collect dominate offsets from image, adding in default 0,0 offset '''
    res = dominant_offsets_helper(offsets)
    res = np.array(sorted(res, key = lambda x: x[-1])[::-1])[:K,0:2]
    return np.concatenate((np.array([[0,0]]),res - [im.shape[0] + 1,im.shape[1] + 1]))
```
